```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("arashnic/book-recommendation-dataset")

print("Path to dataset files:", path)
```

```python
import numpy as np
import pandas as pd
```

```python
books = pd.read_csv(path + "/Books.csv")
```

```python
reatings = pd.read_csv(path + "/Ratings.csv")
```

```python
users = pd.read_csv(path + "/Users.csv")
```

```python
books.head()
```

```python
reatings.head()
```

```python
reatings_with_books = reatings.merge(books, on="ISBN")
```

```python
reatings_with_books
```

```python
num_of_reating_per_book = reatings_with_books.groupby('ISBN').count()['Book-Rating'].reset_index()
num_of_reating_per_book.rename(columns={'Book-Rating':'num_of_reating'},inplace=True)
num_of_reating_per_book
```

```python
avg_of_reating_per_book = reatings_with_books.groupby('ISBN')['Book-Rating'].mean().reset_index()
avg_of_reating_per_book.rename(columns={'Book-Rating':'avg_of_reating'},inplace=True)
avg_of_reating_per_book
```

```python
merged_df = pd.merge(books, num_of_reating_per_book, on='ISBN')
merged_df = pd.merge(merged_df, avg_of_reating_per_book, on='ISBN')
```

```python
reatings
```

```python
final_df = pd.merge(reatings, merged_df, on='ISBN')
```

```python
final_df.head()
```

```python
x = final_df.groupby('User-ID').count()['Book-Rating'] > 200
knowladgeable_users = x[x].index
```

```python
filtered_books = final_df[final_df['User-ID'].isin(knowladgeable_users)]
```

```python
filtered_books
```

```python
y = filtered_books.groupby('Book-Title').count()['num_of_reating'] >= 50
famous_books = y[y].index
```

```python
final_df_2 = filtered_books[filtered_books['Book-Title'].isin(famous_books)]
```

```python
final_df_2 = final_df_2.reset_index(drop=True)
```

```python
final_df_2
```

```python
final_pt_df = final_df_2.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
final_pt_df.fillna(0, inplace=True)
```

```python
final_pt_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(final_pt_df)
```

```python
def recommend(book_name):
  recommended_books = []
  index = np.where(final_pt_df.index == book_name)[0][0]
  distance = sorted(list(enumerate(similarity[index])), key=lambda x:x[1], reverse=True)[1:6]
  for i in distance:
    recommended_books.append(final_pt_df.index[i[0]])
  return recommended_books
```

```python
def recommendation(book_name):
  recommended_books_name_list = recommend(book_name=book_name)
  recommended_books = []
  for name in recommended_books_name_list:
    filtering_by_book_name = final_df_2[final_df_2['Book-Title'] == name]
    removing_dublicates = filtering_by_book_name.drop_duplicates('ISBN')
    picking_the_best_book = removing_dublicates.sort_values(['num_of_reating' and 'Book-Rating' and 'avg_of_reating'], ascending=False).head(1).to_dict(orient='records')
    recommended_books.append(picking_the_best_book[0])
  converting_into_dataframe = pd.DataFrame(recommended_books)
  return converting_into_dataframe
```
